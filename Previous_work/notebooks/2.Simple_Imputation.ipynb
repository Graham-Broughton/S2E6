```python
from tqdm import tqdm
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

import warnings
warnings.filterwarnings("ignore")

from drive.MyDrive.Kaggle.June_2022_na_imputation.src.functions import *
```

```python
data = pd.read_csv('drive/MyDrive/Kaggle/June_2022_na_imputation/src/data.csv', index_col='row_id')
sample = pd.read_csv('drive/MyDrive/Kaggle/June_2022_na_imputation/src/sample_submission.csv', index_col='row-col')
```

```python
col_list, F1, F2, F3, F4, missing_cols = get_lists(data)
```

```python
strategies = ['mean', 'median']
```

```python
df_dict = {}
for strat in strategies:
    se = SimpleImputer(strategy=strat, missing_values=np.nan)
    transformed = se.fit_transform(data)
    df_dict[strat] = pd.DataFrame(transformed, columns=se.feature_names_in_)
```
