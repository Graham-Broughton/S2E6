```python
df = pd.DataFrame()
for run in F4_list:
    for (fold, data) in pkl[run]:
        print(data)
        folds = pd.DataFrame(data)
        folds.rename({'loss': f'{run}_loss_{fold}', 'val_loss': f'{run}_val_loss_{fold}', 'rmse': f'{run}_rmse_{fold}', 'val_rmse': f'{run}_val_rmse_{fold}', 'lr': f'{run}_lr_{fold}'}, axis=1, inplace=True)
        df = pd.concat([df, folds], axis=1)
```
